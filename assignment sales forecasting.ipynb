```python
import statsmodels.api as sm
```

```python
y1 = data1['historyamount'].resample('MS').mean()
y2 = data2['historyamount'].resample('MS').mean()
y3 = data3['historyamount'].resample('MS').mean()
y4 = data4['historyamount'].resample('MS').mean()
y5 = data5['historyamount'].resample('MS').mean()
```

```python
type(y1)
```

```python
x1 =y1.to_frame()
x2 =y2.to_frame()
x3 =y3.to_frame()
x4 =y4.to_frame()
x5 =y5.to_frame()
```

```python
x1.head()
```

```python
model_1 = sm.tsa.statespace.SARIMAX(x1['historyamount'],order = (1,0,1),seasonal_order = (1,0,1,12))
results = model_1.fit()
```

```python
x1['forecast'] = results.predict(start = 20,end = 38,dynamic = True)
x1[['historyamount','forecast']].plot(figsize = (12,8))
```

```python
print('rmse value for model_1 is :',sqrt(mean_squared_error(x1['forecast'][20:],x1['historyamount'][20:])))
```
